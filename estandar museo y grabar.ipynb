cod_area tiene: FLOAT y Nan

```python
df['cod_area'] = df['cod_area'].fillna(0)
```

```python
df['cod_area'] = df['cod_area'].round(decimals=0)
```

```python
df['cod_area'] = df['cod_area'].astype(int)
```

```python
df['cod_area'] = df['cod_area'].astype(str)
```

```python
df['numero de telefono'] = df['cod_area'] + df['telefono']
```
