```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('consolidado.csv')

# Tomamos la base de datos ya consolidada como dataframe
```

```python
# Creamos una base de datos SQL temporal, alojada en la memoria
temp_db = create_engine('sqlite:///:memory:')
```

```python
df.to_sql(name='tabla',con=temp_db)

# Nombramos la tabla principal como "tabla"
```

```python
newdf = pd.read_sql(sql='tabla',con=temp_db)

# Leemos si la tabla esta bien
```

```python
newdf
```

```python
result = pd.read_sql_query(sql='SELECT cod_localidad,id_provincia,id_departamento,categoria,provincia,localidad,nombre,domicilio,mail,web FROM tabla',con=temp_db)

# Nos quedamos con las columnas importantes
```
