```python
df['espacio_INCAA'] = df['espacio_INCAA'].fillna(0)
```

```python
df.loc[df.espacio_INCAA=='si','espacio_INCAA']='1'
df.loc[df.espacio_INCAA=='Si','espacio_INCAA']='1'
df.loc[df.espacio_INCAA=='SI','espacio_INCAA']='1'
df.loc[df.espacio_INCAA=='sI','espacio_INCAA']='1'
```

```python
df['espacio_INCAA'] = df['espacio_INCAA'].astype(int)
```

```python
pd.pivot_table(df,index='Provincia',aggfunc='sum',values=['Pantallas','Butacas','espacio_INCAA']).sort_values(ascending=False,by='Pantallas')
```

Con esto respondemos a lo requerido de CINES, por provincia, ordenando de manera descendente sobre 'pantallas'
